```python
from sklearn.datasets import make_blobs, make_classification
import pylab as plt, numpy as np
```

```python
n_samples = 3000
centers = np.mgrid[0:10, 0:2][0] # each class
X, y = make_blobs(n_samples=n_samples, centers=centers, shuffle=False)#, random_state=42)
```

```python
# other feature
y_ = np.random.random(y.shape)
```

```python
plt.scatter(X[:,0],X[:,1]); plt.show()
plt.scatter(X[:,0],X[:,1], c=y); plt.show()
plt.scatter(X[:,0],X[:,1], c=y_); plt.show()
```

```python
# include paradox
t = np.tan(np.radians(130))
shear_x = np.array(((1, t), (0, 1))).T
# restructure dataset
X_ = X.copy()
for i in np.unique(y):
    # center
    c = X[y==i].mean(0)
    # mvt
    X_[y==i] -= c
    # rot
    X_[y==i] = X_[y==i].dot(shear_x)
    # remvt
    X_[y==i] += c
```

```python
plt.scatter(X_[:,0],X_[:,1]); plt.show()
plt.scatter(X_[:,0],X_[:,1], c=y); plt.show()
plt.scatter(X_[:,0],X_[:,1], c=y_); plt.show()
```
